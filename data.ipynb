```python
import cv2 
import os 
import time
import uuid 
```

```python
data_dir = 'D:\sign-lang\My-Data\signdata'
```

```python
labels = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z'] 
num_img = 220
```

```python
for label in labels:
    label_dir = os.path.join(data_dir, label)
    os.makedirs(label_dir, exist_ok=True) #creats directory is doesn't exist
    print(f"Directory created : {label_dir}")

    cap = cv2.VideoCapture(0)
    print('Get Ready, will start collecting images in 15 seconds for {}'.format(label))
    time.sleep(15) #before starting sleeps for 15 seconds
    print('Collecting images for {}'.format(label))
    print("Press 'q' to stop collecting for {}".format(label))

    for img_num in range(num_img):
        ret, frame = cap.read()
        img_path = os.path.join(label_dir, f'{label}.{uuid.uuid1()}.jpg')
        cv2.imwrite(img_path, frame)

        cv2.imshow('frame', frame)
        time.sleep(1) # takes imgae in every 1 second

        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Exiting the loop as 'q' was pressed.")
            break

    cap.release()
    cv2.destroyAllWindows()

    print(f"Finished collecting images for {label}")
```
